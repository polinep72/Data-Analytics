Удалить строчки с пустыми значениями

```python
df = df.dropna()
df
```

Дублирующиеся строки

```python
df.duplicated().describe() # посмотрим есть ли дубликаты
```

```python
df[df.duplicated()] # посмотрим дубликаты
```

```python
df=df.drop(df[df.duplicated()].index, axis=0)# удаление дубликатов
```

```python
df.info()
```

Переведем данные столбца PuFrom к одному виду

```python
df["PuFrom"]=df["PuFrom"].str.upper()
df["PuFrom"]
```

Разделим время поездки на два столбца минуты и секунды

```python
df['Time'] = pd.to_datetime(df.Time) # переведем столбей Time в тип datetime
df['hour'] = df['Time'].dt.hour
df['minute'] = df['Time'].dt.minute
df
```

```python
df.info()
```

```python
df["Time, min"]= round((df["hour"] + (df["minute"]/60)), 2)
df
```

Удалим ненужные столбцы

```python
df= df.drop(['Time', 'hour', 'minute'], axis=1)
df
```
