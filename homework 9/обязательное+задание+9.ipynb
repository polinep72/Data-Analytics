Получить из таблицы студентов их ФИО и количество предметов

```python
sqlQuery = """
  SELECT students.surname|| ' ' || students.name as ФИО, faculty.quantity_group as 'Количество предметов'   FROM students
  JOIN groups ON students.group_id = groups.id
  JOIN faculty ON groups.faculty_id = faculty.id
"""
df = pd.read_sql(sqlQuery, con)
df
```

Получить количество студентов каждого факультета

```python
sqlQuery = """
  SELECT  faculty.name as 'Наименование факультета', count(students.id) as 'Количество студентов'  FROM students
  JOIN groups ON students.group_id = groups.id
  JOIN faculty ON groups.faculty_id = faculty.id
  GROUP BY faculty.id
"""
df = pd.read_sql(sqlQuery, con)
df
```
